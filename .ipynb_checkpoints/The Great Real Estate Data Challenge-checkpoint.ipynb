### Outlier Detection

```python
df3[(df3['carpet_area']/df3['num_rooms'])<300]
```

#### Remove outliers in price_per_carpet_area: rows that are outside 1 standard deviation(I assume that the feature has values that are normally distributed)

```python
df3['price_per_carpet_area'].describe()
```

```python
def remove_outliers_ppca(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('Locality'):
        m =np.mean(subdf.price_per_carpet_area)
        st=np.std(subdf.price_per_carpet_area)
        reduced_df= subdf[(subdf.price_per_carpet_area>(m-st))&(subdf.price_per_carpet_area<=(m+st))]
        df_out = pd.concat([df_out,reduced_df], ignore_index=True)
    return df_out
```

```python
df4 = remove_outliers_ppca(df3)
df4.shape
```

```python
df4['num_rooms'].unique()
```

```python
def plot_scatter_chart(df,Locality):
    room2 = df[(df.Locality == Locality) & (df.num_rooms == 2)]
    room3 = df[(df.Locality == Locality) & (df.num_rooms == 3)]
    matplotlib.rcParams['figure.figsize'] = (15,10)
    plt.scatter(room2['carpet_area'],room2['Sale Price'], marker='*',color='blue', label= '2 Bedrooms', s=50)
    plt.scatter(room3['carpet_area'],room3['Sale Price'], marker='*',color='green', label= '3 Bedrooms', s=50)
    plt.xlabel('Total Square foot area')
    plt.ylabel('Price')
    plt.title(Locality)
    plt.legend()
```

 ### Algorithm selection - Comparing Linear Regression with DecissionTree Regressor and Lasso to find out which one best suits the problem at hand

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'linear_regression' : {
            'model': Pipeline([
                ('scaler', StandardScaler()),
                ('regressor', LinearRegression())
            ]),
            'params': {
                'regressor__normalize': [True, False]
            }
        },
        'lasso': {
            'model': Pipeline([
                ('scaler', StandardScaler()),
                ('regressor', Lasso())
            ]),
            'params': {
                'regressor__alpha': [1,2],
                'regressor__selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': Pipeline([
                ('scaler', StandardScaler()),
                ('regressor', DecisionTreeRegressor())
            ]),
            'params': {
                'regressor__criterion' : ['mse','friedman_mse'],
                'regressor__splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs = GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X, y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores, columns=['model','best_score','best_params'])

# Load the dataset
X_train,X_test,y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)

# Find the best model
best_model = find_best_model_using_gridsearchcv(X_train, y_train)

# Train and evaluate the best model on the test data
model = best_model.loc[0, 'model']
model.fit(X_train, y_train)
model.score(X_test,y_test)
```
